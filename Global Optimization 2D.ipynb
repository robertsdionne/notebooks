```python
%matplotlib inline
import math
import matplotlib
from matplotlib import pyplot as p
import numpy as np
import random
```

```python
def Ackley(x0, x1):
    n, a, b, c = 2.0, 20.0, 0.2, 2.0 * math.pi
    s1, s2 = x0**2 + x1**2, np.cos(c * x0) + np.cos(c * x1)
    return -a * np.exp(-b * np.sqrt(1.0 / n * s1)) - np.exp(1.0 / n * s2) + a + math.e
```

```python
def Rosenbrock(x0, x1):
    return 100.0 * (x1 - x0**2)**2 + (x0 - 1)**2
```

```python
def Deviation(f):
    return lambda alpha, x0, x1: alpha - np.minimum(alpha, f(x0, x1))
```

```python
V = Deviation(Ackley)
```

```python
def Estimator(V, alpha, N, a, b, c, d):
    samples0 = (b - a) * np.random.random_sample(N) + a
    samples1 = (d - c) * np.random.random_sample(N) + c
    return (b - a) * (d - c) / float(N) * V(alpha, samples0, samples1).sum()
```

```python
# Binary search over the range of f to find \alpha^*
def FindEssentialInfimum(f, N, a, b, c, d, epsilon):
    V = Deviation(f)
    delta = 1.0
    alpha_max = f((b - a) * np.random.random_sample() + a, (d - c) * np.random.random_sample() + c)
    alpha = alpha_max - delta
    alpha_min = -float('inf')
    expand = True
    while expand:
        integral = Estimator(V, alpha, N, a, b, c, d)
        if integral > 0:
            delta *= 2.0
            alpha_max = alpha
            alpha = alpha - delta
        else:
            delta /= 2.0
            alpha_min = alpha
            alpha = alpha + delta
            expand = False
    while alpha_max - alpha_min > epsilon:
        integral = Estimator(V, alpha, N, a, b, c, d)
        if integral > 0:
            alpha_max = alpha
        else:
            alpha_min = alpha
        alpha = (alpha_min + alpha_max) / 2.0
    return alpha
```

```python
FindEssentialInfimum(Ackley, 2**16, -15.0, 30.0, -15.0, 30.0, 1e-10)
```

```python
# Binary search over parameter x to find the subdomain containing the global optimum coordinate x^* and value \alpha^*
def FindMinimizingParameter(f, N, a, b, c, d, epsilon):
    x0_max = b
    x0_min = a
    x1_max = d
    x1_min = c
    while x0_max - x0_min > epsilon or x1_max - x1_max > epsilon:
        alpha_star = float('inf')
        print
        print u'locating x0\u207a within', (x0_min, x0_max)
        while True:
            x0 = (x0_min + x0_max) / 2.0
            potential_alpha_star = FindEssentialInfimum(f, N, x0_min, x0_max, x1_min, x1_max, epsilon)
            if potential_alpha_star < alpha_star:
                alpha_star = potential_alpha_star
            else:
                break
            print u'\u03b1\u207a', alpha_star
            alpha_left = FindEssentialInfimum(f, N, x0_min, x0, x1_min, x1_max, epsilon)
            alpha_right = FindEssentialInfimum(f, N, x0, x0_max, x1_min, x1_max, epsilon)
            if abs(alpha_star - alpha_left) < abs(alpha_star - alpha_right):
                x0_max = x0
                print 'x0 in', (x0_min, x0_max), 'dx0', x0_max - x0_min, u'd\u03b1', abs(alpha_star - alpha_left)
            else:
                x0_min = x0
                print 'x0 in', (x0_min, x0_max), 'dx0', x0_max - x0_min, u'd\u03b1', abs(alpha_star - alpha_right)
        alpha_star = float('inf')
        print
        print u'locating x1\u207a within', (x1_min, x1_max)
        while True:
            x1 = (x1_min + x1_max) / 2.0
            potential_alpha_star = FindEssentialInfimum(f, N, x0_min, x0_max, x1_min, x1_max, epsilon)
            if potential_alpha_star < alpha_star:
                alpha_star = potential_alpha_star
            else:
                break
            print u'\u03b1\u207a', alpha_star
            alpha_left = FindEssentialInfimum(f, N, x0_min, x0_max, x1_min, x1, epsilon)
            alpha_right = FindEssentialInfimum(f, N, x0_min, x0_max, x1, x1_max, epsilon)
            if abs(alpha_star - alpha_left) < abs(alpha_star - alpha_right):
                x1_max = x1
                print 'x1 in', (x1_min, x1_max), 'dx1', x1_max - x1_min, u'd\u03b1', abs(alpha_star - alpha_left)
            else:
                x1_min = x1
                print 'x1 in', (x1_min, x1_max), 'dx1', x1_max - x1_min, u'd\u03b1', abs(alpha_star - alpha_right)
    return ((x0_min, x0_max), (x1_min, x1_max))
```
